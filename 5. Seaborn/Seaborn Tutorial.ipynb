## Import seaborn

* Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures. Seaborn helps you explore and understand your data.

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Correlation with Heatmap

df.corr()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
sns.heatmap(df.corr())
```

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair Plot
### A pair plot is also known as scatterplot, in which one variable in the same data row is matched with
* another varaible's value, like this:
 Pairs plots are just elaborations on this, showing all varaibles paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
df['smoker'].unique()
```

```python
sns.pairplot(df,hue='smoker')
```

```python
# Dist plot
#Dist plot helps us to check the distribution of the columns feature
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
# Categorical features
# Violinplot()
# Count plot()
# Boxplot()
```

```python
df
```

```python
# count plot

sns.countplot(y='day', data=df)
```

```python
# bar plot

sns.barplot(x='total_bill', y='sex', data=df)
```

```python
# Box plot

sns.boxplot('smoker', 'total_bill', data=df)
```

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill',y="day",hue="smoker", data=df)
```

```python
# Violin plot
# helps us to see both distribution of data in terms of kernel density estimation and the box plot
```

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
