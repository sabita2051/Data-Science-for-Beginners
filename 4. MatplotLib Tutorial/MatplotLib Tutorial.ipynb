## Import matplotlib

* Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. 

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# %matplotlib inline sets the backend of matplotlib to the 'inline' backend: With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it. The resulting plots will then also be stored in the notebook document.
```

```python
import numpy as np
```

```python
#Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##Plotting using matplotlib

##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig("Test.png")
```

```python
y=x*x
```

```python
# plt plot

plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")


# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# Subplot()
# Two rows and one column
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active

plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin, 'r--')
plt.plot('sine')

# Set the second subplot as active, and make the second subplot

plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g--')
plt.title('cosine')

# Show the figure
plt.show()

```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,)
plt.bar(x2,y2,color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
# Histograms

a = np.array([22,87,5,43,56,25,43,20,31,5,51,27])
plt.hist(a)
plt.title("histogram")
plt.show()
```

```python
# Box plot using Matplotlib

data = [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangular box plot
#plt.boxplot(data,vert=False,patch_artist=False);
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# Data to plot in piechart

labels = 'python', 'c++', 'java','Ruby'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellow', 'lightcoral', 'lightskyblue']
explode = (0.5, 0,0,0) # explode 1st slice

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
